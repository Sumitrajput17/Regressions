```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,4,7,9,13])
y = np.array([5,20,23,45,65,78,99])
```

```python
plt.figure(figsize=(7,6))
plt.xlabel('Years of Experiences')
plt.ylabel('Salary in $USD')
plt.title('Year vs Salary')
plt.scatter(x,y)
plt.legend(['train_data'])
plt.show()
```

```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = x.reshape(-1,1)
model.fit(x,y)
```

```python
predicted_value = model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,predicted_value,color='red')
plt.show()
```

```python
new_x =[[265]]
print('pridicted value :',model.predict(new_x))
```

```python
plt.figure(figsize=(7,6))
plt.xlabel('Years of Experiences')
plt.ylabel('Salary in $USD')
plt.title('Year vs Salary prediction line')
plt.scatter(x,y)
plt.plot(x,predicted_value,color='red')
plt.legend(['train_data','best_fit_line'])
plt.show()#this is same part as above
```

```python
model.intercept_
```

```python
years = 8.5
print('predicted_value =',model.predict([[years]]))
```

```python
plt.figure(figsize=(7,6))
plt.xlabel('Years of Experiences')
plt.ylabel('Salary in $USD')
plt.title('Year vs Salary prediction line')
plt.scatter(x,y,color = 'red')
plt.plot(x,predicted_value,color='lightblue')
x_test = np.array([8.5,15])
plt.scatter(x_test,model.predict(x_test.reshape(-1,1)),color = 'darkgreen')
plt.legend(['train_data','best_fit_line','test_data'])
plt.show()
```
